```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
# Load the image
nii_filename = 'sub-01_T1w.nii.gz'
img = nib.load(nii_filename)
data = np.array(img.get_fdata())
data.shape
```

```python
# save data to disk as a npy file
np.save('sub-01_T1w.npy', data)
```

```python
# Show an axial slice
slice = data[:,:,160]
plt.imshow(slice, cmap='gray')
plt.colorbar()

# change the color limits to see the brain better
plt.imshow(slice, cmap='gray', clim=[0,2000])
```

```python
# Show a sagittal slice
slice = data[104,:,:]
slice = np.rot90(slice)
plt.imshow(slice, cmap='gray')
plt.colorbar()
# change the color limits to see the brain better
plt.imshow(slice, cmap='gray', clim=[0,1000])
```

```python
# show a coronal slice
slice = data[:,104,:]
slice = np.rot90(slice)
plt.imshow(slice, cmap='gray')
plt.colorbar()
# change the color limits to see the brain better
plt.imshow(slice, cmap='gray', clim=[0,1500])
```

### Part 3: Installing and importing libraries
In most cases, the code that we want to use is contained in external libraries. To be able to access those libraries, we need to *import* them into our notebook. The cell below shows you how to import three common libraries: numpy, pandas, and matplotlib. Yay!

```python
# Import libraries with standard abbreviations
import numpy as np      # For numerical calculations
import pandas as pd     # For data handling
import matplotlib.pyplot as plt  # For plotting

# Create some sample ECG data
time = np.linspace(0, 1, 100)  # Create 100 time points from 0 to 1 second
ecg_signal = np.sin(2 * np.pi * 10 * time)  # Simulate ECG-like wave

# Plot the ECG signal
plt.figure(figsize=(10, 4))
plt.plot(time, ecg_signal)
plt.title('Simulated ECG Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude (mV)')
plt.grid(True)
plt.show()

```

### Part 4: Working with simulated data

```python
# Create a simple patient dataset
# We will define a "dictionary" that stores medical records for 5 patients
data = {
    'Patient_ID': [1, 2, 3, 4, 5],
    'Age': [25, 45, 65, 35, 55],
    'Heart_Rate': [72, 78, 82, 70, 75],
    'Blood_Pressure_Systolic': [120, 135, 142, 118, 128],
    'Blood_Pressure_Diastolic': [80, 85, 88, 75, 82]
}

# Create a DataFrame
# We will now convert our dictionary into a "DataFrame", which is a tabular data structure that is convenient to work with
df = pd.DataFrame(data)
print("\nPatient Data:")
print(df)

# Calculate average heart rate
# This is an example of calling a "function", which is a piece of code that is meant to be used many times over.
# We will call the mean() function and print out its output.
avg_hr = df['Heart_Rate'].mean()
print("\nAverage Heart Rate:", round(avg_hr, 1), "bpm")

# Create a bar plot of heart rates
# The library matplotlib contains easy-to-use functions for making bar plots. Fire!
plt.figure(figsize=(8, 4))
plt.bar(df['Patient_ID'], df['Heart_Rate'])
plt.title('Patient Heart Rates')
plt.xlabel('Patient ID')
plt.ylabel('Heart Rate (bpm)')
plt.show()

```
